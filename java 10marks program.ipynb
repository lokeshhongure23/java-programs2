### 1.Write a Java program to display all the alphabets between ‘A’ to *Z” after every 2 seconds.

```python
import java.io.*;
public class demo extends Thread
{
    public static void main(String args[])
    {
        demo t = new demo();
        t.start();
    }
    public void run()
    {
        try{
            for(char ch='A';ch<= 'Z';ch++){
                System.out.println(ch);
                sleep(2000);
            }
        }
        catch(Exception e){
            System.out.println(e);
        }
    }
}
```

### 4.	Write a Java Program to create the hash table that will maintain the mobile number and student name. Display the details of students using the Enumeration interface.	

```python
import java.util.*;
class Hash{
    public static void main(String args[]){
        Hashtable<String,String>ht=new Hashtable<String,String>();
        ht.put("name:","Lokesh");
        ht.put("mob no:","123456789");
        Enumeration<String>e=ht.keys();
        while(e.hasMoreElements()){
            String key=e.nextElement();
            System.out.println(key+ht.get(key));
        }
    }
}
```

### 6. Write a JSP program to check whether a given number is prime or not. Display the result in red color.

```python
# html file
<html>
<head>
<title>prime no </title>
</head>
<body>
<form action = "prime.jsp" method = "post">
Number : <input type = "number" name="num" required><br><br>
<input type = "submit" value = "check prime">
</form>
</body>
</html>    
```

```python
#jsp file
<%@ page language="java" import="java.util.*"%>
<html>
<head>
<title>prime </title>
</head>
<body text="red">
<%
try{
    int x = Integer.parseInt(request.getParameter("num"));
    int i, flag = 1;

    for(i = 2; i < x; i++){
        if(x % i == 0){
            flag = 0;
            break;
        }
    }

    if(flag == 1){
        out.println("number is prime");
    }
    else{
        out.println("number is not prime");
    }

}
catch(Exception e){
    out.println(e);
}
%>
</body>
</html>
```

### Q.7	Write a JSP program that accepts an integer number from the user through an HTML form, calculates its factorial, and displays the result on the browser.	

```python
#html
<html>
<head>
<title>facorial calculator </title>
</head>
<body>
<h2> factorial calculator</h2>
<form action = "fact.jsp" method = "post">
Number : <input type = "number" name="num" required><br><br>
<input type = "submit" value = "calculate">
</form>
</body>
</html>
```

```python
#jsp
<%@ page language ="java" import="java.util.*"%>
<html>
<head>
<title>factorial</title>
</head>
<body>
<%
try{
    int x = Integer.parseInt(request.getParameter("num"));
    int i,fact = 1;
    for(int i=1 ; i<=x ; i++){
        fact = fact*i;
    }
    out.println("Number :"+x);
    out.println("factorial :"+fact);
}
catch(Exception e){
    out.println(e);
}
%>
</body>
</html>
```

### 9. Write a JSP program to check whether given number is Perfect or not. (Use Include directive).	[10M]

```python
#html
<html>
<head>
<title> Perfect </title>
</head>
<body>
<h2> perfect </h2>
<form action = "perfect.jsp" method = "post">
Enter Number : <input type = "number" name="num" required><br><br>
<input type = "submit" value = "check">
</form>
</body>
</html>
```

```python
#jsp
<%@ page language ="java" import = "java.util.*"%>
<html>
<head>
<title> perfect no </title>
</head>
<body>
<%
try{
    int x = Integer.parseInt(request.getParameter("num"));
    int i,sum =0;
    for(i=1;i<=x/2;i++){
        if(x%i==0){
            sum = sum+i;
        }
    
    }
    if(sum==x){
        out.println("number is perfect");
    }
    else{
        out.println("number is not perfect");
    }
}
catch(Exception e){
    out.println(e);
}
%>
</body>
</html>
```
