### 1.Write a Java program to display all the alphabets between ‘A’ to *Z” after every 2 seconds.

```python
import java.io.*;
public class demo extends Thread
{
    public static void main(String args[])
    {
        demo t = new demo();
        t.start();
    }
    public void run()
    {
        try{
            for(char ch='A';ch<= 'Z';ch++){
                System.out.println(ch);
                sleep(2000);
            }
        }
        catch(Exception e){
            System.out.println(e);
        }
    }
}
```

### 2. Write a java program to read ‘N’ names of your friends, store it into HashSet and display them in ascending order.

```python
import java.util.*;

public class FriendNames {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);

       
        HashSet<String> friends = new HashSet<>();

        System.out.print("Enter number of friends: ");
        int n = sc.nextInt();
        sc.nextLine(); 

        
        System.out.println("Enter friend names:");
        for (int i = 0; i < n; i++) {
            String name = sc.nextLine();
            friends.add(name); 
        }

        
        TreeSet<String> sortedFriends = new TreeSet<>(friends);

        
        System.out.println("\nFriends in ascending order:");
        for (String name : sortedFriends) {
            System.out.println(name);
        }

        sc.close();
    }
}
```

### 4.	Write a Java Program to create the hash table that will maintain the mobile number and student name. Display the details of students using the Enumeration interface.	

```python
import java.util.*;
class Hash{
    public static void main(String args[]){
        Hashtable<String,String>ht=new Hashtable<String,String>();
        ht.put("name:","Lokesh");
        ht.put("mob no:","123456789");
        Enumeration<String>e=ht.keys();
        while(e.hasMoreElements()){
            String key=e.nextElement();
            System.out.println(key+ht.get(key));
        }
    }
}
```

```python
#servlet
import java.io.*;
import javax.servlet.*;
import javax.servlet.http.*;
import javax.servlet.annotation.WebServlet;
@WebServlet("/add")
public class AddServlet extends HttpServlet {
protected void doPost(HttpServletRequest request, HttpServletResponse response)
throws ServletException, IOException {
int num1 = Integer.parseInt(request.getParameter("num1"));
int num2 = Integer.parseInt(request.getParameter("num2"));
int sum = num1 + num2;
response.setContentType("text/html");
PrintWriter out = response.getWriter();
out.println("<h2>Sum: " + sum + "</h2>");
out.close();
}
}
```

### 6. Write a JSP program to check whether a given number is prime or not. Display the result in red color.

```python
# html file
<html>
<head>
<title>prime no </title>
</head>
<body>
<form action = "prime.jsp" method = "post">
Number : <input type = "number" name="num" required><br><br>
<input type = "submit" value = "check prime">
</form>
</body>
</html>    
```

```python
#jsp file
<%@ page language="java" import="java.util.*"%>
<html>
<head>
<title>prime </title>
</head>
<body text="red">
<%
try{
    int x = Integer.parseInt(request.getParameter("num"));
    int i, flag = 1;

    for(i = 2; i < x; i++){
        if(x % i == 0){
            flag = 0;
            break;
        }
    }

    if(flag == 1){
        out.println("number is prime");
    }
    else{
        out.println("number is not prime");
    }

}
catch(Exception e){
    out.println(e);
}
%>
</body>
</html>
```

### 10. Write a java program to accept ‘N’ integers from a user store them into LinkedList Collection and display only negative integers.

```python
import java.util.*;
import java.util.LinkedList;
public class NegativeIntegersList {
    public static void main(String[] args) {
        Scanner sc = new Scanner(System.in);
        LinkedList<Integer> list = new LinkedList<>();
        System.out.print("Enter number of integers: ");
        int n = sc.nextInt();
        System.out.println("Enter " + n + " integers:");
        for(int i = 0; i < n; i++) {
            int num = sc.nextInt();
            list.add(num);
        }
        System.out.println("Negative integers are:");
        for(int num : list) {
            if(num < 0) {
                System.out.print(num + " ");
            }
        }
        
        sc.close();
    }
}
```
